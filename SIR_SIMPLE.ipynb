```python
import numpy as np 
import matplotlib.pyplot as plt
from scipy.integrate import odeint

#function that returns the differential equations
def du_dt(u,t):
    return[-alpha*u[0]*u[1],alpha*u[0]*u[1]-mu*u[1],mu*u[1]]

#initial conditions
s=int(input('Number of Susceptible Individuals: '))
i=int(input('Number of Infected Individuals: '))
r=int(input('Number of Recovered Individuals: '))
u0=[s,i,r]
alpha= float(input('Base rate for S-->I: '))
mu= float(input('Base rate for I-->R: '))

#time interval
ts=np.arange(0,100,1)
 #solution
us=odeint(du_dt,u0,ts)
#assigning the three solutions 
S,I,R=us[:,0],us[:,1],us[:,2]
#figure shape
fig=plt.subplots(figsize=(12,10))

plt.plot(ts,S,linewidth=2.5,label='Susceptible')
plt.plot(ts,I,linewidth=2.5,label='Infected')
plt.plot(ts,R,linewidth=2.5,label='Recovered')
# plt.grid()
plt.legend(fontsize=20)

plt.xlabel('Days',fontsize=20)
plt.ylabel('Population',fontsize=20)

plt.show
```
